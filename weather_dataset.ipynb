```python
df['Date'] = pd.to_datetime(df['Date'])
print("Date column converted to datetime.")
print(df['Date'].head())
```

```python
print("Missing values in each column:")
print(df.isnull().sum())
```

```python
df['Month'] = df['Date'].dt.month
print("Month column extracted:")
print(df['Month'].head())
```

```python
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Summer'
    elif month in [6, 7, 8, 9]:
        return 'Monsoon'
    else:
        return 'Post-Monsoon'

df['Season'] = df['Month'].apply(get_season)
print("Season column created:")
print(df[['Date', 'Month', 'Season']].head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 7))
sns.lineplot(x='Date', y='Temperature_Avg (°C)', data=df)
plt.title('Daily Average Temperature Trend')
plt.xlabel('Date')
plt.ylabel('Average Temperature (°C)')
plt.grid(True)
plt.show()
```

```python
df['Year'] = df['Date'].dt.year
monthly_avg_temp = df.groupby(['Year', 'Month'])['Temperature_Avg (°C)'].mean().reset_index()

plt.figure(figsize=(15, 7))
sns.lineplot(x='Month', y='Temperature_Avg (°C)', hue='Year', data=monthly_avg_temp, marker='o')
plt.title('Monthly Average Temperature Trend (2024-2025)')
plt.xlabel('Month')
plt.ylabel('Average Temperature (°C)')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.legend(title='Year')
plt.show()
```

```python
seasonal_avg_rainfall = df.groupby('Season')['Rainfall (mm)'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Season', y='Rainfall (mm)', data=seasonal_avg_rainfall, palette='viridis')
plt.title('Average Rainfall per Season')
plt.xlabel('Season')
plt.ylabel('Average Rainfall (mm)')
plt.grid(axis='y')
plt.show()
```

```python
seasonal_avg_rainfall = df.groupby('Season')['Rainfall (mm)'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Season', y='Rainfall (mm)', hue='Season', data=seasonal_avg_rainfall, palette='viridis', legend=False)
plt.title('Average Rainfall per Season')
plt.xlabel('Season')
plt.ylabel('Average Rainfall (mm)')
plt.grid(axis='y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Get unique seasons
unique_seasons = df['Season'].unique()

# Determine the number of subplots needed
num_seasons = len(unique_seasons)
n_cols = 2  # Number of columns for subplots
n_rows = (num_seasons + n_cols - 1) // n_cols # Calculate rows needed

plt.figure(figsize=(n_cols * 7, n_rows * 6)) # Adjust figure size dynamically

for i, season in enumerate(unique_seasons):
    plt.subplot(n_rows, n_cols, i + 1)

    # Filter data for the current season
    season_df = df[df['Season'] == season]

    # Calculate AQI category distribution
    aqi_category_counts = season_df['AQI_Category'].value_counts()

    # Create pie chart
    plt.pie(aqi_category_counts, labels=aqi_category_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
    plt.title(f'AQI Category Distribution in {season}')
    plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.

plt.tight_layout()
plt.show()
```
